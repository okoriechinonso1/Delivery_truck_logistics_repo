```python
# Combining ontime and delay attributes into a single column with name 'ontime/delay'

raw_dF['onetime/delay'] = raw_dF.ontime.replace({np.NaN, 'G'}, {0, 1})
```

```python
raw_dF['vehicle_states'] = [x[:2] for x in raw_dF['vehicle_no'].astype(str)]
raw_dF["Origin_states"] = [x[-1] for x in raw_dF['Origin_Location'].str.split(',')]

raw_dF['Dest_states'] = raw_dF['Destination_Location'].str.split(',').str[-1]
```

```python
raw_dF['vehicle_states'] = raw_dF['vehicle_states'].replace(('tn', 'hr'), ('TN', 'HR'))

def lower_case_converter(row): # using function to convert each element of the Origin_states feature into lowercase
    row['Origin_states'] = row['Origin_states'].capitalize()
    return row['Origin_states']

raw_dF['Origin_states'] = raw_dF.apply(lower_case_converter, axis=1)


def lower_case_convert(row): # using function to convert each element of the Dest_states feature into lowercase
    row['Dest_states'] = row['Dest_states'].capitalize()
    return row['Dest_states']

raw_dF['Dest_states'] = raw_dF.apply(lower_case_convert, axis=1)
```

```python
def check_mis_label(column): # Using function definition to check for mis-labels and spurious data
    unique_values = column.unique()
    return unique_values

print(check_mis_label(raw_dF['GpsProvider']), '\n')
#check_mis_label(raw_dF['Market/Regular'])
print(check_mis_label(raw_dF['vehicle_no']), "\n")
print(check_mis_label(raw_dF['Origin_Location']), "\n")
print(check_mis_label(raw_dF['Destination_Location']), "\n")
print(check_mis_label(raw_dF['Current_Location']), "\n")
print(check_mis_label(raw_dF['Destination_Location']), "\n")
print(check_mis_label(raw_dF['OriginLocation_Code']), "\n")
print(check_mis_label(raw_dF['DestinationLocation_Code']), "\n")
print(check_mis_label(raw_dF['customerID']), "\n")
print(check_mis_label(raw_dF['supplierNameCode']), "\n")
print(check_mis_label(raw_dF['Material Shipped']))
```

```python
!pip install geopy
import geopy
from geopy import distance

# finding the distance between origin and desitination
from geopy.distance import geodesic as GD

km_distances = []

for _, row in raw_dF.iterrows():
    km_distances.append(GD(row['Org_lat_lon'], row['Des_lat_lon']).km)

raw_dF['Org_Dest_distance'] = km_distances

```

### Treating Missing values


```python
for col in raw_dF.columns:
    if raw_dF[col].isna().sum() > 0:
        print(col, raw_dF[col].isna().mean().round(4)*100)
```

```python
# Filling missing values in drivers attribute with unknown
raw_dF['Driver_Name'] = raw_dF['Driver_Name'].fillna('Unknown')

raw_dF['TRANSPORTATION_DISTANCE_IN_KM'] = raw_dF['TRANSPORTATION_DISTANCE_IN_KM'].rolling(min_periods=1, center=True, window=10).mean()

raw_dF['Driver_MobileNo'] = raw_dF['Driver_MobileNo'].fillna('Unknown')

raw_dF['vehicleType'] = raw_dF['vehicleType'].fillna('Unknown')

raw_dF['actual_eta'] = raw_dF['actual_eta'].fillna(method="bfill")

raw_dF["Minimum_kms_to_be_covered_in_a_day"] = raw_dF["Minimum_kms_to_be_covered_in_a_day"].fillna(method='bfill')

raw_dF["GpsProvider"] = raw_dF["GpsProvider"].fillna('Unknown')

raw_dF["Data_Ping_time"] = raw_dF["Data_Ping_time"].fillna(method="bfill")
```

```python
# Some variables have thesame null values in thesame samples.  The features are
# dependent on each other and so the sample rows from such attributes will be dropped

raw_dF.dropna(axis=0, how='any', subset=['Current_Location'], inplace=True)
```

```python
# Imputing values into rows of trip_end_date feature where there are missing values

import datetime
import random

df_subdata = raw_dF[raw_dF['trip_end_date'].isna()]

for i in df_subdata.index:
    if df_subdata['ontime/delay'][i] == 0:
        df_subdata['trip_end_date'][i] = df_subdata['actual_eta'][i]
    else:
        df_subdata['trip_end_date'][i] = df_subdata['Planned_ETA'][i] - datetime.timedelta(random.randint(0,3))

raw_dF = pd.concat([raw_dF, df_subdata], axis=0)

#Removing duplicates since concatenation of the two dataframe 
raw_dF.dropna(subset=['trip_end_date'], inplace=True)
```
